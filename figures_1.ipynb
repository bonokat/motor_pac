```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.signal import butter, filtfilt

# Function to create bandpass-filtered noise
def bandpass_filter(data, lowcut, highcut, fs, order=4):
    nyquist = 0.5 * fs
    low = lowcut / nyquist
    high = highcut / nyquist
    b, a = butter(order, [low, high], btype='band')
    y = filtfilt(b, a, data)
    return y

# Simulation parameters
sampling_rate = 1000  # in Hz
time = np.arange(0, 2, 1/sampling_rate)  # 2 seconds of data
noise = np.random.randn(len(time))  # Random noise

# Frequency ranges for each brainwave band
theta_range = (4, 8)  # Theta: 4-8 Hz
alpha_range = (8, 12)  # Alpha: 8-12 Hz
beta_range = (12, 30)  # Beta: 12-30 Hz
gamma_range = (30, 100)  # Gamma: 30-100 Hz

# Generate complex oscillations using bandpass-filtered noise
theta_wave = bandpass_filter(noise, theta_range[0], theta_range[1], sampling_rate)
alpha_wave = bandpass_filter(noise, alpha_range[0], alpha_range[1], sampling_rate)
beta_wave = bandpass_filter(noise, beta_range[0], beta_range[1], sampling_rate)
gamma_wave = bandpass_filter(noise, gamma_range[0], gamma_range[1], sampling_rate)

# Create a figure with 4 subplots
fig, axs = plt.subplots(4, 1, figsize=(10, 8), sharex=True)
# fig.suptitle('Brain oscillations in bands', fontsize=16)

# Plot each brainwave oscillation
axs[0].plot(time, theta_wave, color='b')
axs[0].set_title('Theta (4-8 Hz)')
axs[0].set_ylabel('Amplitude')

axs[1].plot(time, alpha_wave, color='g')
axs[1].set_title('Alpha (8-12 Hz)')
axs[1].set_ylabel('Amplitude')

axs[2].plot(time, beta_wave, color='r')
axs[2].set_title('Beta (12-30 Hz)')
axs[2].set_ylabel('Amplitude')

axs[3].plot(time, gamma_wave, color='purple')
axs[3].set_title('Gamma (30-100 Hz)')
axs[3].set_xlabel('Time (s)')
axs[3].set_ylabel('Amplitude')

# Adjust layout
plt.tight_layout(rect=[0, 0, 1, 0.96])

# Show the plot
plt.show()

```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.signal import butter, filtfilt

# Function to create bandpass-filtered noise
def bandpass_filter(data, lowcut, highcut, fs, order=4):
    nyquist = 0.5 * fs
    low = lowcut / nyquist
    high = highcut / nyquist
    b, a = butter(order, [low, high], btype='band')
    y = filtfilt(b, a, data)
    return y

# Simulation parameters
sampling_rate = 1000  # in Hz
time = np.arange(0, 10, 1/sampling_rate)  # 10 seconds of data
noise = np.random.randn(len(time))  # Base noise signal

# Generate bandpass-filtered signals for different frequency bands
delta_band = bandpass_filter(noise, 0.5, 4, sampling_rate) * 0.5
theta_band = bandpass_filter(noise, 4, 8, sampling_rate) * 0.3
alpha_band = bandpass_filter(noise, 8, 12, sampling_rate) * 0.2
beta_band = bandpass_filter(noise, 12, 30, sampling_rate) * 0.15
gamma_band = bandpass_filter(noise, 30, 100, sampling_rate) * 0.05

# Combine all components to create a realistic EEG signal
eeg_signal = delta_band + theta_band + alpha_band + beta_band + gamma_band

# Add amplitude modulation to some frequency bands
theta_amplitude_mod = 1 + 0.5 * np.sin(2 * np.pi * 0.05 * time)  # Slow modulation
eeg_signal += theta_amplitude_mod * theta_band

alpha_amplitude_mod = 1 + 0.4 * np.sin(2 * np.pi * 0.1 * time)
eeg_signal += alpha_amplitude_mod * alpha_band

# Add more transient events (e.g., sleep spindles or K-complexes)
spindle_time1 = (time > 2) & (time < 2.5)  # First spindle occurrence
spindle1 = 0.3 * np.sin(2 * np.pi * 14 * time[spindle_time1])  # 14 Hz spindle
eeg_signal[spindle_time1] += spindle1

spindle_time2 = (time > 6) & (time < 6.5)  # Second spindle occurrence
spindle2 = 0.3 * np.sin(2 * np.pi * 12 * time[spindle_time2])  # 12 Hz spindle
eeg_signal[spindle_time2] += spindle2

# Add random sharp transients to simulate artifacts or sharp waveforms
sharp_transient_time = (time > 8) & (time < 8.1)
sharp_transient = 1.0 * np.exp(-50 * (time[sharp_transient_time] - 8)**2)
eeg_signal[sharp_transient_time] += sharp_transient

# Add physiological noise
physiological_noise = 0.05 * np.random.randn(len(time))
eeg_signal += physiological_noise

# Plot the simulated EEG signal
plt.figure(figsize=(12, 6))
plt.plot(time, eeg_signal, color='b', linewidth=1)
plt.title('Physiologically Inspired Simulated EEG Signal (0.5-100 Hz)')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.grid(True)
plt.tight_layout()

# Show the plot
plt.show()
```

```python
sns.set_style("darkgrid")

# Generate data for two sine waves with a frequency of 5 Hz
frequency = 3
sampling_rate = 1000  # Samples per second
duration = 1.0  # Seconds

# Create the time array
t = np.linspace(0, duration, int(sampling_rate * duration), endpoint=False)

# Create two sine waves with different phase shifts
amplitude = .5
phase_shift = np.pi/4  # 45 degrees in radians

y1 = amplitude * np.sin(2 * np.pi * frequency * t)
y2 = amplitude * np.sin(2 * np.pi * frequency * t + phase_shift)

# Plot both sine waves in one figure
plt.plot(t, y1, label=f"f = {frequency} Hz, ϕ = 0\N{DEGREE SIGN}", color='#fa0581') #fa0581 #1ecbe1
plt.plot(t, y2, label=f"f = {frequency} Hz, ϕ = 45\N{DEGREE SIGN}", color='#05fa7e') #05fa7e #e1341e

plt.ylim([-1, 1])
plt.xlim([-.1, 1.2])

# Add labels and legend
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.legend()

# Add title
title = 'Two Sine Waves with Phase Shift'
plt.title(title)

# Add grid
plt.grid(True, linestyle='--', alpha=0.7)

# Show the plot
plt.show()

# Save the figure
plt.savefig(fname=f'{title}.jpg', dpi=300, format='jpg')
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

def amplitude_function(t):
    return 1 + 0.95 * np.sin(np.pi * 4 * t)  # Varying amplitude with a frequency of 4 Hz

# Define time vector
t = np.linspace(0, 2, 1000)  # 2 seconds, 1000 samples

# Subplot 1: 4 Hz Sinusoid
freq_4hz = 4
signal_4hz = np.sin(np.pi * freq_4hz * t)

plt.figure(figsize=(12, 10))

plt.subplot(3, 1, 1)
plt.plot(t, signal_4hz, color='#C83F69')
plt.title('Theta: 4 Hz', fontsize=30)
plt.grid(False)
plt.axis('off')

# Subplot 2: 40 Hz Sinusoid with varying amplitude
freq_40hz = 40
amplitude_40hz = amplitude_function(t)  # Varying amplitude
signal_40hz = amplitude_40hz * np.sin(np.pi * freq_40hz * t)

plt.subplot(3, 1, 2)
plt.plot(t, signal_40hz, color='#8BC34A')
plt.title('Gamma: 40 Hz', fontsize=30)
plt.grid(False)
plt.axis('off')

# Subplot 3: Phase-Amplitude Coupling
# Combine the two signals with coupling
coupled_signal = signal_4hz + 0.5 * signal_40hz

plt.subplot(3, 1, 3)
plt.plot(t, coupled_signal, color='#5A3FC8')
plt.title('Theta-gamma PAC', fontsize=30)
plt.grid(False)
plt.axis('off')

plt.tight_layout()
plt.show()
```
